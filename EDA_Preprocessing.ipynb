```python
from google.colab import files
uploaded = files.upload()

import pandas as pd

file_name = list(uploaded.keys())[0]

df = pd.read_csv(file_name)

print("File loaded successfully:", file_name)
df.head()

print(df.describe())

print("\nGender:\n", df['Gender'].value_counts())
print("\nInternship Mode:\n", df['Internship_Mode'].value_counts())
print("\nDepartment:\n", df['Department'].value_counts())

df['Task_Completion_%'] = (df['Tasks_Completed'] / df['Tasks_Assigned']) * 100

df['Performance_Level'] = df['Final_Performance_Rating'].apply(
    lambda x: 'High' if x >= 4 else ('Medium' if x == 3 else 'Low')
)


df.head()

df.to_csv("intern_dataset.csv", index=False)

from google.colab import files
files.download("intern_dataset.csv")
```
